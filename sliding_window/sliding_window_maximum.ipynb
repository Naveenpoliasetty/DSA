## 239. Sliding Window Maximum   <span style="color : red"> Hard </span>

You are given an array of integers nums, there is a sliding window of size k which is moving from the very left of the array to the very right. You can only see the k numbers in the window.<br> Each time the sliding window moves right by one position.

Return the max sliding window.


Example 1: <br>

Input: nums = [1,3,-1,-3,5,3,6,7], k = 3 <br>
Output: [3,3,5,5,6,7] <br>
Explanation:  <br>
Window position                Max <br>
---------------               ----- <br>
[1  3  -1] -3  5  3  6  7       3 <br>
 1 [3  -1  -3] 5  3  6  7       3 <br>
 1  3 [-1  -3  5] 3  6  7       5 <br>
 1  3  -1 [-3  5  3] 6  7       5 <br>
 1  3  -1  -3 [5  3  6] 7       6 <br>
 1  3  -1  -3  5 [3  6  7]      7 <br>

<br>
Example 2:

Input: nums = [1], k = 1
Output: [1]

### Brute force solution

```python
from collections import deque
from typing import List
import heapq
class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        ans = []
        for i in range(len(nums)-k + 1):
            ans.append(max(nums[i:i+k]))
        return ans
    
s = Solution()

nums = [1,3,-1,-3,5,3,6,7]
k = 3

s.maxSlidingWindow(nums, k)
```

### Better solution

```python
from collections import deque
from typing import List
import heapq
class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        heap = []
        ans = []

        for i in range(len(nums)):
            heapq.heappush(heap, (-nums[i], i))
            if i >= k-1:
                while heap and heap[0][1] <= i - k:
                    heapq.heappop(heap)
                
                ans.append(-heap[0][0])

        return ans
    

s = Solution()

nums = [1,3,-1,-3,5,3,6,7]
k = 3

s.maxSlidingWindow(nums, k)
```

### Optimal answer

```python
from collections import deque
import heapq
class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        dq = deque()
        ans = []

        for i in range(len(nums)):

            while dq and dq[0] <= i-k:
                dq.popleft()

            while dq and nums[dq[-1]] <= nums[i]:
                dq.pop()
            dq.append(i)

            if i >= k - 1:
                ans.append(nums[dq[0]])

        return ans
    

s = Solution()

nums = [1,3,-1,-3,5,3,6,7]
k = 3

s.maxSlidingWindow(nums, k)
```
