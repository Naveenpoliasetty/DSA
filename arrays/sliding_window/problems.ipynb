## 239. Sliding Window Maximum   <span style="color : red"> Hard </span>

You are given an array of integers nums, there is a sliding window of size k which is moving from the very left of the array to the very right. You can only see the k numbers in the window.<br> Each time the sliding window moves right by one position.

Return the max sliding window.


Example 1: <br>

Input: nums = [1,3,-1,-3,5,3,6,7], k = 3 <br>
Output: [3,3,5,5,6,7] <br>
Explanation:  <br>
Window position                Max <br>
---------------               ----- <br>
[1  3  -1] -3  5  3  6  7       3 <br>
 1 [3  -1  -3] 5  3  6  7       3 <br>
 1  3 [-1  -3  5] 3  6  7       5 <br>
 1  3  -1 [-3  5  3] 6  7       5 <br>
 1  3  -1  -3 [5  3  6] 7       6 <br>
 1  3  -1  -3  5 [3  6  7]      7 <br>

<br>
Example 2:

Input: nums = [1], k = 1
Output: [1]

### Brute force solution

```python
from collections import deque
from typing import List
import heapq
class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        ans = []
        for i in range(len(nums)-k + 1):
            ans.append(max(nums[i:i+k]))
        return ans
    
s = Solution()

nums = [1,3,-1,-3,5,3,6,7]
k = 3

s.maxSlidingWindow(nums, k)
```

### Better solution

```python
from collections import deque
from typing import List
import heapq
class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        heap = []
        ans = []

        for i in range(len(nums)):
            heapq.heappush(heap, (-nums[i], i))
            if i >= k-1:
                while heap and heap[0][1] <= i - k:
                    heapq.heappop(heap)
                
                ans.append(-heap[0][0])

        return ans
    

s = Solution()

nums = [1,3,-1,-3,5,3,6,7]
k = 3

s.maxSlidingWindow(nums, k)
```

### Optimal answer

```python
from collections import deque
import heapq
class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        dq = deque()
        ans = []

        for i in range(len(nums)):

            while dq and dq[0] <= i-k:
                dq.popleft()

            while dq and nums[dq[-1]] <= nums[i]:
                dq.pop()
            dq.append(i)

            if i >= k - 1:
                ans.append(nums[dq[0]])

        return ans
    

s = Solution()

nums = [1,3,-1,-3,5,3,6,7]
k = 3

s.maxSlidingWindow(nums, k)
```

## 1438. Longest Continuous Subarray With Absolute Diff Less Than or Equal to Limit ;  <span style="color : yellow">      Medium </span>

Given an array of integers nums and an integer limit, return the size of the longest non-empty subarray such that the absolute difference between any two elements of this subarray is less than or equal to limit.

 

Example 1:<br>

Input: nums = [8,2,4,7], limit = 4<br> <br>
Output: 2 
Explanation: All subarrays are: <br>
[8] with maximum absolute diff |8-8| = 0 <= 4.<br>
[8,2] with maximum absolute diff |8-2| = 6 > 4. <br>
[8,2,4] with maximum absolute diff |8-2| = 6 > 4.<br>
[8,2,4,7] with maximum absolute diff |8-2| = 6 > 4.<br>
[2] with maximum absolute diff |2-2| = 0 <= 4.<br>
[2,4] with maximum absolute diff |2-4| = 2 <= 4.<br>
[2,4,7] with maximum absolute diff |2-7| = 5 > 4.<br>
[4] with maximum absolute diff |4-4| = 0 <= 4.<br>
[4,7] with maximum absolute diff |4-7| = 3 <= 4.<br>
[7] with maximum absolute diff |7-7| = 0 <= 4. <br>
Therefore, the size of the longest subarray is 2.<br> <br>
Example 2:<br>

Input: nums = [10,1,2,4,7,2], limit = 5<br>
Output: 4 <br>
Explanation: The subarray [2,4,7,2] is the longest since the maximum absolute diff is |2-7| = 5 <= 5.
Example 3:<br>
<br>
Input: nums = [4,2,2,2,4,4,2,2], limit = 0<br>
Output: 3<br>

```python
class Solution:
    def longestSubarray(self, nums: List[int], limit: int) -> int:
        i, j = 0, 0
        ans = []
        for i in range(len(nums)):
            min_val = float('inf')
            j = i
            while j < len(nums):
                min_val = min(min_val, nums[j])
                if nums[i]-min_val <= limit:
                        ans.append(nums[i])
                j += 1
        return ans


s = Solution()
nums = [8, 2, 4, 7]
limit = 4

print(s.longestSubarray(nums, limit))
```

## 1438. Longest Continuous Subarray With Absolute Diff Less Than or Equal to Limit ;  <span style="color : yellow">      Medium </span>

 
Given an array of integers nums and an integer limit, return the size of the longest non-empty subarray such that the absolute difference between any two elements of this subarray is less than or equal to limit. <br>

 

Example 1: <br>

Input: nums = [8,2,4,7], limit = 4 <br>
Output: 2  <br>
Explanation: All subarrays are:  <br>
[8] with maximum absolute diff |8-8| = 0 <= 4. <br>
[8,2] with maximum absolute diff |8-2| = 6 > 4.  <br>
[8,2,4] with maximum absolute diff |8-2| = 6 > 4. <br>
[8,2,4,7] with maximum absolute diff |8-2| = 6 > 4. <br>
[2] with maximum absolute diff |2-2| = 0 <= 4. <br>
[2,4] with maximum absolute diff |2-4| = 2 <= 4. <br>
[2,4,7] with maximum absolute diff |2-7| = 5 > 4. <br>
[4] with maximum absolute diff |4-4| = 0 <= 4. <br>
[4,7] with maximum absolute diff |4-7| = 3 <= 4. <br>
[7] with maximum absolute diff |7-7| = 0 <= 4.  <br>
Therefore, the size of the longest subarray is 2. <br> br>
Example 2: <br>
 <br>
Input: nums = [10,1,2,4,7,2], limit = 5
Output: 4  <br>
Explanation: The subarray [2,4,7,2] is the longest since the maximum absolute diff is |2-7| = 5 <= 5. <br> <br>
Example 3: <br>

Input: nums = [4,2,2,2,4,4,2,2], limit = 0 <br>
Output: 3 <br>

```python
from collections import deque
class Solution:
    def longestSubarray(self, nums: List[int], limit: int) -> int:
        max_len = 0
        min_dq = deque()
        max_dq = deque()

        i = 0

        for j in range(len(nums)):
            while max_dq and max_dq[-1] < nums[j]:
                max_dq.pop()
            max_dq.append(nums[j])

            while min_dq and min_dq[-1] > nums[j]:
                min_dq.pop()
            min_dq.append(nums[j])

            while max_dq[0] - min_dq[0] > limit:
                if max_dq[0] == nums[i]:
                    max_dq.popleft()
                if min_dq[0] == nums[i]:
                    min_dq.popleft()
                i += 1
            max_len = max(max_len, j - i + 1)
        return max_len


        
```
