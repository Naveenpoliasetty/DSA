Example 2:

Input: ["we","say",":","yes"]

0 <= strs.length < 100
0 <= strs[i].length < 200
strs[i] contains only UTF-8 characters.

```python
from typing import List
def encode(strs: List[str]) -> str:
    s = None
    for word in strs:
        s += len(word) + word + "#"
    return s
s = encode(["Naveen", "Shyam", "Ajay"])
print(s)
```
